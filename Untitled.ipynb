```python
import pandas as pd
import matplotlib.pyplot as plt
```

## importing the dataset

```python
df = pd.read_csv('./data/netflix_titles.csv')
```

```python
df.head(5)
```

```python
#missing values in each colums
df.isnull().sum()
```

## See total and percentage of missing values

```python
missing = df.isnull().sum()
percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing, 'Percent': percent})
print(missing_df[missing_df['Missing Values'] > 0])
```

## 🧹 2. Clean or Handle Each Column

## A. 'director' & 'cast'

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
```

## B. 'country'

```python
df['country'] = df['country'].fillna('Unknown')
```

## C. 'date_added'

```python
df['date_added']=df['date_added'].fillna(pd.to_datetime('1900-01-01')) 
```

## D. 'rating'


```python
df['rating']= df['rating'].fillna('Not Rated')
```

## E. 'duration'

```python
# Create new columns based on type
df['duration'] = df['duration'].fillna('0')

df['duration_int'] = df['duration'].str.extract(r'(\d+)').astype(int)

df['duration_type'] = df['duration'].str.extract('([a-zA-Z ]+)', expand=False).str.strip()
```

## Remove Duplicates (if any)

```python
df.drop_duplicates(inplace=True)
```

```python
df.count()
```

```python
df.head(5)
```

```python
# Check original 'country' values with missing ones
df['country'].value_counts(dropna=False).head(10)
```

```python
df.to_csv('./data/netflix_titles_cleaned.csv', index=False)
```

## importing cleaned data set

```python
cleaned_df = pd.read_csv("./data/netflix_titles_cleaned.csv")
```

## Count of Movies vs TV Shows



```python
type_counts = cleaned_df['type'].value_counts()
```

```python
type_counts
```

```python
plt.figure(figsize=(6,4))
type_counts.plot(kind='bar', color=['#1DB954', '#FF5733'])
plt.title('Count of Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
for i, value in enumerate(type_counts):
    plt.text(i, value + 0, str(value), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Top 10 Countries by Content

```python
top_countries = df['country'].value_counts().head(10)
top_countries
```

```python
plt.figure(figsize=(8,5))
top_countries.plot(kind='barh', color='skyblue')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
for i, value in enumerate(top_countries):
    plt.text(value + 0, i, str(value), va='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

##  Content Added Over Time

```python
cleaned_df['date_added'] = pd.to_datetime(cleaned_df['date_added'], errors='coerce')
```

```python
cleaned_df['year_added'] = cleaned_df['date_added'].dt.year
```

```python
print(cleaned_df[['date_added', 'year_added']].head())
```

## Count titles per year and sort chronologically

```python
added_per_year = cleaned_df['year_added'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.plot(added_per_year.index, added_per_year.values, marker='o', color='purple')
plt.title('Number of Titles Added to Netflix per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)

# Add number labels on the line
for x, y in zip(added_per_year.index, added_per_year.values):
    plt.text(x, y + 20, str(int(y)), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

## Most Common Ratings


```python
rating_counts = cleaned_df['rating'].value_counts().head(10)

plt.figure(figsize=(7,4))
rating_counts.plot(kind='bar', color='orange')
plt.title('Top 10 Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
for i, value in enumerate(rating_counts):
    plt.text(i, value + 0, str(value), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Most Popular Genres (Categories)

```python
from collections import Counter
```

```python
all_genres = cleaned_df['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in all_genres for genre in sublist]
genre_counts = pd.Series(Counter(flat_genres)).sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
genre_counts.plot(kind='barh', color='teal')
plt.title('Top 10 Netflix Genres')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
for i, value in enumerate(genre_counts):
    plt.text(value + 10, i, str(value), va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Trend of Release Years

```python
# Count titles per release year
release_trend = cleaned_df['release_year'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.plot(release_trend.index, release_trend.values, color='darkblue', marker='o')
plt.title('Trend of Content Release Years on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)

for x, y in zip(release_trend.index, release_trend.values):
    if y > 200:  # avoid cluttering early low years
        plt.text(x, y + 20, str(y), ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```

 ## Distribution of Ratings

```python
rating_distribution = cleaned_df['rating'].value_counts()

plt.figure(figsize=(10,5))
rating_distribution.plot(kind='bar', color='coral')
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)

# Add value labels
for i, value in enumerate(rating_distribution):
    plt.text(i, value + 50, str(value), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Average Movie Duration vs TV Show Seasons

```python
# Filter and calculate mean duration
# Create numeric duration column from 'duration'
cleaned_df['duration_int'] = cleaned_df['duration'].str.extract(r'(\d+)').astype(float)
avg_movie_duration = cleaned_df[cleaned_df['type'] == 'Movie']['duration_int'].mean()
avg_tv_seasons = cleaned_df[cleaned_df['type'] == 'TV Show']['duration_int'].mean()

# Create comparison bar chart
plt.figure(figsize=(6,4))
plt.bar(['Movies (min)', 'TV Shows (seasons)'], [avg_movie_duration, avg_tv_seasons], color=['#1DB954', '#FF5733'])

plt.title('Average Movie Duration vs TV Show Seasons')
plt.ylabel('Average Duration')

# Add value labels
plt.text(0, avg_movie_duration + 1, f"{avg_movie_duration:.1f}", ha='center', fontweight='bold')
plt.text(1, avg_tv_seasons + 0.1, f"{avg_tv_seasons:.1f}", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
